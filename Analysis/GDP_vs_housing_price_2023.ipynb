```python
import pandas as pd
```

```python
home_df = pd.read_csv('../Source/Metro_Home_Values.csv')
home_df.head()
```

```python
# Define the columns to keep:
# First five columns
columns_to_keep = home_df.columns[:5]

# Extract year from columns and identify columns for 2022, 2023, and 2024
def extract_year_from_column(col_name):
    try:
        return col_name.split('-')[0]  # Extract year from 'YYYY-MM-DD'
    except IndexError:
        return None

# Create a set of years to keep
years_to_keep = {'2023', '2024'}

# Filter columns to keep only those starting with the specified years
columns_for_years = [col for col in home_df.columns if extract_year_from_column(col) in years_to_keep]

# Combine the columns to keep
all_columns_to_keep = list(columns_to_keep) + columns_for_years

# Filter DataFrame
home_df_filtered = home_df[all_columns_to_keep]
home_df_filtered
```

```python
df_sorted = home_df_filtered.sort_values(by='StateName')
df_sorted.drop(['RegionType', 'SizeRank', 'RegionID'], axis = 1, inplace = True)
# List of columns to keep
columns_to_keep = ['RegionName', 'StateName', '2023-12-31']
df_all = df_sorted[columns_to_keep]
df_all = df_all.rename(columns={'2023-12-31': 'Prices in 2023'})
df_all = df_all.groupby('StateName')['Prices in 2023'].mean().reset_index()
df_all.head()
```

```python
gdp_df_all = pd.read_csv('../Source/GDP_2023.csv')
gdp_df_all.drop('GeoFips', axis = 1, inplace = True)
gdp_df_all = gdp_df_all.rename(columns={'GeoName': 'StateName'})
gdp_df_all = gdp_df_all.rename(columns={'2023': 'GDP 2023'})
gdp_df_all = gdp_df_all.iloc[1:].reset_index(drop=True)
gdp_df_all.head()
```

```python
state_mapping = {
    'Alabama': 'AL', 'Alaska': 'AK', 'Arizona': 'AZ', 'Arkansas': 'AR', 'California': 'CA',
    'Colorado': 'CO', 'Connecticut': 'CT', 'Delaware': 'DE', 'Florida': 'FL', 'Georgia': 'GA',
    'Hawaii': 'HI', 'Idaho': 'ID', 'Illinois': 'IL', 'Indiana': 'IN', 'Iowa': 'IA',
    'Kansas': 'KS', 'Kentucky': 'KY', 'Louisiana': 'LA', 'Maine': 'ME', 'Maryland': 'MD',
    'Massachusetts': 'MA', 'Michigan': 'MI', 'Minnesota': 'MN', 'Mississippi': 'MS', 'Missouri': 'MO',
    'Montana': 'MT', 'Nebraska': 'NE', 'Nevada': 'NV', 'New Hampshire': 'NH', 'New Jersey': 'NJ',
    'New Mexico': 'NM', 'New York': 'NY', 'North Carolina': 'NC', 'North Dakota': 'ND', 'Ohio': 'OH',
    'Oklahoma': 'OK', 'Oregon': 'OR', 'Pennsylvania': 'PA', 'Rhode Island': 'RI', 'South Carolina': 'SC',
    'South Dakota': 'SD', 'Tennessee': 'TN', 'Texas': 'TX', 'Utah': 'UT', 'Vermont': 'VT',
    'Virginia': 'VA', 'Washington': 'WA', 'West Virginia': 'WV', 'Wisconsin': 'WI', 'Wyoming': 'WY'
}

gdp_df_all['StateName'] = gdp_df_all['StateName'].map(state_mapping)
merged_df = pd.merge(df_all, gdp_df_all, on='StateName')
df_sorted_gdp = merged_df.sort_values(by='GDP 2023')
df_sorted_gdp.head()
```

```python
compare_df = df_sorted_gdp.set_index('StateName', inplace = False)
compare_df.head()
```

```python
states_to_compare = ['VT', 'CA', 'HI']
rows = compare_df.loc[states_to_compare]
print(rows)
```

```python
sorted_compare_df = compare_df.sort_values(by='Prices in 2023', ascending=True)
sorted_compare_df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for seaborn
sns.set(style="whitegrid")

# Create a figure with two subplots
fig, axes = plt.subplots(2, 1, figsize=(10, 12), sharex=True)

# Plot the bar graph for State vs. Prices
sns.barplot(x='StateName', y='Prices in 2023', data=df_sorted_gdp, ax=axes[1], palette="Blues_d", hue = None)
axes[1].set_title('State vs. Prices in 2023')
axes[1].set_ylabel('Prices in 2023')

# # Plot the bar graph for State vs. GDP
sns.barplot(x='StateName', y='GDP 2023', data=df_sorted_gdp, ax=axes[0], palette="Greens_d", hue = None)
axes[0].set_title('State vs. GDP in 2023')
axes[0].set_ylabel('GDP in 2023 (in Trillions)')

# Rotate x-axis labels for better readability
plt.xticks(rotation=90)

# Adjust layout to prevent overlap
plt.tight_layout()

# Set background color for each subplot's axes
for ax in axes:
    ax.set_facecolor('#CCCCCC')  # Set the background color of the plot area
# Set background color for the entire figure
plt.gcf().set_facecolor('#CCCCCC')  # Set the background color of the figure

# Save figure
plt.savefig("../IMAGES/GDPvsPrices_all.png")

# Show the plots
plt.show()
```

```python
from scipy import stats
gdp = df_sorted_gdp['GDP 2023']
prices = df_sorted_gdp['Prices in 2023']
plt.scatter(gdp,prices)
plt.xlabel("GDP")
plt.ylabel("House Prices")
plt.show()
```

```python
import scipy.stats as st
print(f"The correlation coefficient between GDP and House Prices is {round(st.pearsonr(gdp,prices)[0],2)}")
```

```python
gdp_df = pd.read_csv('../Source/GDP_2023.csv')
gdp_states = ['Arizona', 'California', 'Texas', 'New York', 'Hawaii', 'Nevada', 'Florida', 'Oregon', 'Washington', 'Utah']
gdp_df = gdp_df[gdp_df['GeoName'].isin(gdp_states)]
gdp_df.drop('GeoFips', axis = 1, inplace = True)
gdp_df = gdp_df.rename(columns={'GeoName': 'StateName'})
gdp_df = gdp_df.rename(columns={'2023': 'GDP 2023'})
gdp_df.head(10)
```

```python
state_mapping = {
    'Arizona': 'AZ',
    'California': 'CA',
    'Texas': 'TX',
    'New York': 'NY',
    'Hawaii': 'HI',
    'Nevada': 'NV',
    'Florida': 'FL',
    'Oregon': 'OR',
    'Washington': 'WA',
    'Utah': 'UT'
}

gdp_df['StateName'] = gdp_df['StateName'].map(state_mapping)
merged_df = pd.merge(df_filtered, gdp_df, on='StateName')
df_sorted_gdp = merged_df.sort_values(by='GDP 2023')
df_sorted_gdp.head(10)
```
