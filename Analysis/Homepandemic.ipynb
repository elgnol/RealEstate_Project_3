```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
from sklearn.linear_model import LinearRegression
import numpy as np
```

```python
home_df = pd.read_csv('../Source/filtered_home_df.csv')
home_df.head()
```

```python
rental_df = pd.read_csv('../Source/rental_df.csv')
rental_df.head()
```

```python
# Extract year from columns and identify columns for  2018, 2019, 2020, 2021, and 2022
def extract_year_from_column(col_name):
    try:
        return col_name.split('-')[0]  # Extract year from 'YYYY-MM-DD'
    except IndexError:
        return None
```

```python
# Create a set of years to keep
years_to_keep = {'2018', '2019', '2020', '2021', '2022'}
```

```python
# Filter columns to keep only those starting with the specified years
columns_for_years = [col for col in home_df.columns if extract_year_from_column(col) in years_to_keep]
```

```python
# Combine the columns to keep
columns_to_keep = columns_for_years
```

```python
# Filter DataFrame
home_df_filtered = home_df[columns_to_keep]
home_df_filtered
```

```python
# Export as a json file
home_df_filtered.head(1).to_json('home_output_data.json', orient='records', indent=4)
```

```python
# Filter columns to keep only those starting with the specified years
columns_for_years_r = [col for col in rental_df.columns if extract_year_from_column(col) in years_to_keep]
```

```python
# Combine the columns to keep
columns_to_keep = columns_for_years_r
```

```python
# Filter DataFrame
rental_df_filtered = rental_df[columns_to_keep]
rental_df_filtered
```

```python
# Export as a json file
rental_df_filtered.head(1).to_json('rental_output_data.json', orient='records', indent=4)
```
